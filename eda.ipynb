```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Training Dataset/training datalist.csv')
display(df.head())
```

```python
display(df.dtypes.to_frame().rename(columns={0: "Data Type"}))

display(df.head())

display(df.info())

display(df.shape)

# Display dataset summary statistics
display(df.describe())

# Display count of missing values in each variable
display(df.isnull().sum().to_frame().rename(columns={0: "Count of Missing Values"}))
```

```python
# data cleaning
df['PPD'].fillna(0, inplace=True)
df.dropna(subset=['Voice handicap index - 10'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
print("Unique labels: ", df['Disease category'].unique())

label_counts = df['Disease category'].value_counts()
print("Label counts: ")
print(label_counts)
print("Label weights: ")
print(label_counts / len(df))
```

```python
# Plot the distribution of the labels

# Set the style and color palette
sns.set(style="whitegrid", palette="pastel")

# Create count plot with interactive features
plt.figure(figsize=(14, 6))
ax = sns.countplot(data=df, x='Disease category', order=sorted(df['Disease category'].unique()), 
                   palette="pastel")
ax.set_title('Distribution of Disease category in the Dataset')
ax.set_xlabel('Disease category')
ax.set_ylabel('Count')
ax.set_xticklabels(ax.get_xticklabels())
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x()+p.get_width()/2, p.get_height()),
                ha='center', va='bottom', fontsize=12)
plt.tight_layout()

# Show interactive plot
plt.show()
```

```python
# pie chart
plt.figure(figsize=(14, 6))
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=180)
plt.title('Distribution of Disease category in the Dataset')
plt.tight_layout()
plt.show()
```

```python
# distribution of age
plt.style.use("seaborn")
fig, ax = plt.subplots(figsize=(15,8))
sns.histplot(df["Age"], kde=True, bins=25, color="indianred")
plt.title("Age of the patients", fontsize=20, y=1.02)
ax.set_xlabel("Age",fontsize=15)
```

```python
# group by sex and disease category
plt.style.use("seaborn")

fig, ax = plt.subplots(figsize=(15, 8))
sns.barplot(x="Sex", y="Count", hue="Disease category", data=df.groupby(["Sex", "Disease category"]).size().reset_index(name="Count"), palette="hot")
ax.set_xlabel("Sex", fontsize=14)
# Customize x-axis tick labels
ax.set_xticklabels(['Male', 'Female'], fontsize=12)
# Additional customization options can be added here
plt.title("Disease category group by sex", fontsize=20, y=1.02)

plt.show()
```

```python
plt.style.use("seaborn")
fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(x="Sex", data=df, palette="hot")
plt.title("Distribution of Sex", fontsize=16)
ax.set_xlabel("Sex", fontsize=14)
ax.set_ylabel("Count", fontsize=14)

plt.show()
```

```python
# Group by "Sex" and "Disease category" and calculate count
df_grouped = df.groupby(["Sex", "Disease category"]).size().reset_index(name="Count")

# Calculate the count of each smoking label
sex_counts = df.groupby("Sex").size()
print(sex_counts)

# Add a new column "Percentage" to df_grouped
df_grouped["Percentage"] = df_grouped.apply(lambda row: row["Count"] / sex_counts[row["Sex"]], axis=1)
display(df_grouped.head())
```

```python
# group by smoke percentage and disease category
plt.style.use("seaborn")

fig, ax = plt.subplots(figsize=(15, 8))
sns.barplot(x="Sex", y="Percentage", hue="Disease category", data=df_grouped, palette="hot")
ax.set_xlabel("Sex", fontsize=14)
# Customize x-axis tick labels
ax.set_xticklabels(['Male', 'Female'], fontsize=12)
# Additional customization options can be added here
plt.title("Disease category group by sex percentage", fontsize=20, y=1.02)


plt.show()
```

```python
# group by smoke and disease category
plt.style.use("seaborn")

fig, ax = plt.subplots(figsize=(15, 8))
sns.barplot(x="Smoking", y="Count", hue="Disease category", data=df.groupby(["Smoking", "Disease category"]).size().reset_index(name="Count"), palette="hot")
ax.set_xlabel("Smoking", fontsize=14)
# Customize x-axis tick labels
ax.set_xticklabels(['never', 'past', 'active', 'e-cigarette'], fontsize=12)
# Additional customization options can be added here
ax.legend(loc='upper right', bbox_to_anchor=(1, 1), title="Disease category", fontsize=12)
plt.title("Disease category group by Smoking", fontsize=20, y=1.02)

plt.show()
```

```python
# show the smoking people distribution
plt.style.use("seaborn")
fig, ax = plt.subplots(figsize=(15,8))
sns.histplot(df["Smoking"], bins=25, color="indianred")
plt.title("Smoking of the patients", fontsize=20, y=1.02)
# Set the x-axis tick positions and labels
tick_positions = [0, 1, 2, 3]
tick_labels = ['never', 'past', 'active', 'e-cigarette']
ax.set_xticks(tick_positions)
ax.set_xticklabels(tick_labels, fontsize=12)
ax.set_xlabel("Smoking",fontsize=15)
```

```python
# Group by "Smoking" and "Disease category" and calculate count
df_grouped = df.groupby(["Smoking", "Disease category"]).size().reset_index(name="Count")

# Calculate the count of each smoking label
smoking_counts = df.groupby("Smoking").size()
print(smoking_counts)

# Add a new column "Percentage" to df_grouped
df_grouped["Percentage"] = df_grouped.apply(lambda row: row["Count"] / smoking_counts[row["Smoking"]], axis=1)
display(df_grouped.head())
```

```python
# group by smoke percentage and disease category
plt.style.use("seaborn")

fig, ax = plt.subplots(figsize=(15, 8))
sns.barplot(x="Smoking", y="Percentage", hue="Disease category", data=df_grouped, palette="hot")
ax.set_xlabel("Smoking", fontsize=14)
# Customize x-axis tick labels
ax.set_xticklabels(['never', 'past', 'active', 'e-cigarette'], fontsize=12)
# Additional customization options can be added here
ax.legend(loc='upper right', bbox_to_anchor=(1, 1), title="Disease category", fontsize=12)
plt.title("Disease category group by Smoking percentage", fontsize=20, y=1.02)


plt.show()
```
